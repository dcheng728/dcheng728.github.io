The series expansion is given by 
$$\Phi(\rho, \phi, z) = \sum_{n=1,m=1}^\infty B_{nm} \cos(m \phi) sin(n\pi z / L) I_m(n \pi rho / L)$$
Denote this by 
$$\Phi(\rho, \phi, z) = \sum_{n=1,m=1}^\infty B_{nm} F_{nm}$$
It is found that 
$$B_{nm} = \frac{16 V}{\pi^2 nm} \frac{1}{I_m(\frac{n \pi}{L}b)} (-1)^{(m-1)/2}$$ if $n,m$ are both odd, otherwise $B_{nm} = 0$ 

```python
z,rho,phi = var('z,rho,phi')
n,m,V,L,b = var('n,m,V,L,b')
```

```python
# Formula for the coefficients, note that we have divided by V, because we are interested in V/V_0
B = 16 / (pi^2 * m * n * bessel_I(m,n * pi * b / L)) * (-1)^((m-1)/2)
show(B)
```

```python
# The components of the expansion
F = cos(m*phi) * sin(n * pi * z / L) * bessel_I(m,n * pi * rho / L)
show(F)
```

```python
Phi = 0
for i in range (30):
    for j in range(30):
        Phi +=  B(n = 2* i + 1, m = 2* j+ 1) * F(n = 2 * i + 1, m = 2 * j + 1)

#show(Phi)
```
