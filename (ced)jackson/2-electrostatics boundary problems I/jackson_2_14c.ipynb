```python
import numpy as np
```

```python
# b taken to be 1, 2V/pi taken to be 1
```

```python
def phi(samples):
    return np.arctan(2 * samples[:,0] * np.sin(2 * samples[:,1])) / (1 - samples[:,0]**4)
```

```python
def xy_to_polar(coords):
    polar = np.zeros( coords.shape )
    r = (coords[:,0]**2 + coords[:,1]**2 )**(1/2)
    polar[:,0] = r
    polar[:,1] = np.arccos( coords[:,1] / r)

    return polar
```

```python
plot_sample_xy = np.random.rand(100000,2) - 0.5
plot_sample_polar = xy_to_polar(plot_sample_xy)
plot_evaled = phi(plot_sample_polar)
```

```python
plot_evaled.max()
```

```python
import matplotlib.pyplot as plt
import matplotlib.tri as tri

plot_coord = plot_sample_xy
plot_height = plot_evaled


# ----------
# Tricontour
# ----------
# Directly supply the unordered, irregularly spaced coordinates
# to tricontour.

fig, ax = plt.subplots(figsize=(14, 12))

#ax.tricontour(plot_coord[:,0], plot_coord[:,1], plot_height, levels=5, linewidths=0.5, colors='k')

cntr = ax.tricontourf(plot_coord[:,0], plot_coord[:,1], plot_height, levels=20, cmap="RdBu_r")


#ax.add_patch(plt.Circle((0, 0), 1, color='g', fill=False))

fig.colorbar(cntr, ax=ax)

#ax.plot(test_xz[:,0], test_xz[:,1], 'ko', ms=1)

#ax.set(xlim=(-1, 1), ylim=(-1, 1))
ax.set_title('tricontour (points)' )

#plt.subplots_adjust(hspace=0.5)
plt.axis('equal')
plt.show()
```
